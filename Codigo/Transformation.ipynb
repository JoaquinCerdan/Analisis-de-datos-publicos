```python
formalizados_SF = pd.read_csv('SF/formalizados_SF.csv')

# Eliminamos filas y columnas innecesarias
formalizados_SF = formalizados_SF.drop(columns="Unnamed: 0")
formalizados_SF.drop([889,890], axis=0, inplace=True)

# Pasamos todas las columnas a String, excepto Año
for i in formalizados_SF.columns:
  if i != "Año":
    formalizados_SF[i] = formalizados_SF[i].astype('string')


# Formatear los numeros a float
columnasInt = ['Importe Total Presup. Licit.', 'Importe Adjudicación Sin Impuestos', 'Importe total Adjudicación']
for col in columnasInt:
  formalizados_SF[col] = formalizados_SF[col].str.replace('.','')
  formalizados_SF[col] = formalizados_SF[col].str.replace(',','.')
  formalizados_SF[col] = pd.to_numeric(formalizados_SF[col], errors='coerce')
  for i in range(len(formalizados_SF)):
    if formalizados_SF.iloc[i][col] == 0 or formalizados_SF.iloc[i][col] == 0.0:
      formalizados_SF.at[i,col] = None

  # Aquí rellenaba los valores vacios con -1 pero me estropeaba los análisis
  # for i in range(len(formalizados_SF)):
  #   if pd.isna(formalizados_SF.iloc[i][col]):
  #   # if formalizados_SF.iloc[i][col] == 'nan':
  #     formalizados_SF.at[i,col] = -1

# Pasamos de string a datetime
formalizados_SF['Fecha Formalización Contrato']= pd.to_datetime(formalizados_SF['Fecha Formalización Contrato'], errors='coerce')

formalizados_SF.reset_index(inplace=True, drop=True)

formalizados_SF.to_csv("formalizados.csv", encoding='utf-8')
```

```python
modificados_SF = pd.read_csv('SF/modificados_SF.csv')

# Eliminamos filas y columnas innecesarias
modificados_SF = modificados_SF.drop(columns="Unnamed: 0")

# Pasamos todas las columnas a String, excepto Año
for i in modificados_SF.columns:
  if i != "Año":
    modificados_SF[i] = modificados_SF[i].astype('string')

# Formatear los numeros a float
columnasInt = ['Importe']
for col in columnasInt:
  modificados_SF[col] = modificados_SF[col].str.replace('.','')
  modificados_SF[col] = modificados_SF[col].str.replace(',','.')
  modificados_SF[col] = pd.to_numeric(modificados_SF[col], errors='coerce')
  for i in range(len(modificados_SF)):
    if modificados_SF.iloc[i][col] == 0:
      modificados_SF.iloc[i][col] = None


# Pasamos de string a datetime
columnasDate = ['Fecha Aprobación', 'Fecha Inicio']
for col in columnasDate:
  modificados_SF[col]= pd.to_datetime(modificados_SF[col], errors='coerce')

modificados_SF.drop([40,41,72,73,102,103], axis=0, inplace=True)

modificados_SF.reset_index(inplace=True, drop=True)

modificados_SF.to_csv("modificados.csv", encoding='utf-8')
```

```python
prorrogados_SF = pd.read_csv('SF/prorrogados_SF.csv')

# Eliminamos filas y columnas innecesarias
prorrogados_SF = prorrogados_SF.drop(columns="Unnamed: 0")

# Pasamos todas las columnas a String, excepto Año
for i in prorrogados_SF.columns:
  if i != "Año":
    prorrogados_SF[i] = prorrogados_SF[i].astype('string')

# Formatear los numeros a float
columnasInt = ['Importe']
for col in columnasInt:
  prorrogados_SF[col] = prorrogados_SF[col].str.replace('.','')
  prorrogados_SF[col] = prorrogados_SF[col].str.replace(',','.')
  prorrogados_SF[col] = pd.to_numeric(prorrogados_SF[col], errors='coerce')
  for i in range(len(prorrogados_SF)):
    if prorrogados_SF.iloc[i][col] == 0:
      prorrogados_SF.iloc[i][col] = None

# Pasamos de string a datetime
columnasDate = ['Fecha Aprobación', 'Fecha Inicio', 'Fecha Fin Prórroga']
for col in columnasDate:
  prorrogados_SF[col]= pd.to_datetime(prorrogados_SF[col], errors='coerce')

prorrogados_SF.drop([115,116,136,137], axis=0, inplace=True)

prorrogados_SF.reset_index(inplace=True, drop=True)

prorrogados_SF.to_csv("prorrogados.csv", encoding='utf-8')
```

```python
renunciados_SF = pd.read_csv('SF/renunciados_SF.csv')

# Eliminamos filas y columnas innecesarias
renunciados_SF = renunciados_SF.drop(columns="Unnamed: 0")

# Pasamos todas las columnas a String, excepto Año
for i in renunciados_SF.columns:
  if i != "Año":
    renunciados_SF[i] = renunciados_SF[i].astype('string')

# Formatear los numeros a float
columnasInt = ['Importe']
for col in columnasInt:
  renunciados_SF[col] = renunciados_SF[col].str.replace('.','')
  renunciados_SF[col] = renunciados_SF[col].str.replace(',','.')
  renunciados_SF[col] = pd.to_numeric(renunciados_SF[col], errors='coerce')
  for i in range(len(renunciados_SF)):
    if renunciados_SF.iloc[i][col] == 0:
      renunciados_SF.iloc[i][col] = None

# Caso específico
renunciados_SF.at[7,'Tipo Tramitación'] = 'Urgente'

renunciados_SF.reset_index(inplace=True, drop=True)

renunciados_SF.to_csv("renunciados.csv", encoding='utf-8')
```
